```python
def plot_stats(feature,TARGET,labely,label_rotation=False,horizontal_layout=True):
    temp = kmeans_train[feature].value_counts()
    df1 = pd.DataFrame({feature: temp.index,'Number of contracts': temp.values})

    # Calculate the percentage of target=1 per category value
    cat_perc = kmeans_train[[feature, TARGET]].groupby([feature],as_index=False).mean()
    #cat_perc.sort_values(by=TARGET, ascending=False, inplace=True)
    
    plt.figure(figsize=(4,5), dpi= 80)
    s = sns.barplot(x = feature, y=TARGET, order=cat_perc[feature], data=cat_perc)
    plt.ylabel('Percent of GENDER_M [%]', fontsize=15)
    plt.title(labely);
    
    plt.show();
```
